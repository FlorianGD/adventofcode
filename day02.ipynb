## Part 2
We have to find the numbers that evenly divise each other on each row, and sum the division

```python
def rec_mod(a):
    # a needs to be sorted beforehand
    # This should never happen since we know exactly two terms match
    if len(a) == 1: 
        return 0
    tete = a[0]
    queue = a[1:]
    for b in queue:
        if b % tete == 0:
            return b // tete
    # if no element matched
    return rec_mod(queue)
```

```python
rec_mod(sorted([3, 8, 6, 5]))
```

```python
res = 0
for row in df.iterrows():
    res += rec_mod(sorted((row[1].values)))
```
